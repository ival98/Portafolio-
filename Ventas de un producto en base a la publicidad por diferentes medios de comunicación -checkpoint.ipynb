```python
feature_cols = ["TV", "Radio", "Newspaper"]
```

```python
X = data [feature_cols]
Y = data ["Sales"]
```

```python
estimator = SVR(kernel="linear")
selector = RFE(estimator, n_features_to_select=2, step=1)
selector.fit(X,Y)
```

```python
selector.support_ #True sera la columna que esta mas relacionada con Y
```

```python
selector.ranking_
```

```python
data2 = data [["TV", "Radio","Sales"]]
```

```python
data2.head()
```

```python
filtro = np.random.rand(len(data2)) < 0.8
entrenamiento = data2[filtro]
pruebas = data2[~filtro]
```

```python
entrenamiento.describe
```

```python
from sklearn import linear_model
mlm = linear_model.LinearRegression()
```

```python
train_x = np.asanyarray(entrenamiento[["TV", "Radio"]])
train_y = np.asanyarray(entrenamiento[["Sales"]])
mlm.fit(train_x, train_y)
```

```python
test_x = np.asanyarray(pruebas[["TV", "Radio"]])
test_y = np.asanyarray(pruebas[["Sales"]])
y_pred = mlm.predict(test_x)
```

```python
print("MSE - Error Cuadrado Medio: %.2f" % np.mean((y_pred - test_y)**2))
print("El coeficiente es: ", mlm.coef_)
```

```python
from sklearn.metrics import r2_score
print("R cuadrado o Coeficiente de Determinacion: %.2f" % r2_score(y_pred, test_y))
print("Error medio absoluto: %.2f" % np.mean(np.absolute(y_pred - test_y)))
print("Variance score: %.2f" % mlm.score(test_x, test_y))
```
