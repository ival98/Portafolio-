```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE 
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
hp = pd.read_csv("./Documents/Anaconda/python-ml-course/datasets/House_predicition_Iowa/train.csv")
```

```python
nonnum_cols = hp.select_dtypes(exclude=['float64', 'int64'])
print(nonnum_cols.columns)
```

```python
from sklearn.preprocessing import LabelEncoder

# List with categorical column names
categorical_columns = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual',
       'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
       'SaleType', 'SaleCondition']

# Create a dictionary to store encoders
encoders = {}

# Create an encoder for each categorical column and add it to the dictionary
for column in categorical_columns:
    encoders[column] = LabelEncoder()
    hp[column] = encoders[column].fit_transform(hp[column])
```

```python
# Saleprice is what we want to predict, X is every other columns transformed with numerical values
X = hp[cols]
Y = hp["SalePrice"]
```

```python
# Recursive Feature Elimination (RFE) is used to determine which columns are better to the model based on Y
estimator = SVR(kernel="linear")
selector = RFE(estimator, n_features_to_select=5 , step=1)
selector.fit(X,Y)
```

```python
# They are rated where 1 is the best variable and those that must be taken into account for the final model
feature_select = zip(X, selector.support_, selector.ranking_)
list(feature_select)
```

```python
fcols = ["LotFrontage","OverallQual","YearRemodAdd","TotRmsAbvGrd","FireplaceQu"]
pred = ["SalePrice"]
```

```python
X = hp[fcols]
Y = hp[pred]
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=3)
```

```python
x_trainset, x_testset, y_trainset, y_testset = train_test_split(X,Y, test_size=0.2, random_state=3)
```

```python
rf.fit(x_trainset, y_trainset)
```

```python
y_pred = rf.predict(x_testset)
```

```python
score = rf.score(x_testset, y_testset)
print("PrecisiÃ³n del modelo:", score)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calcular MAE, MSE, y R^2
y_pred = rf.predict(x_testset)
mae = mean_absolute_error(y_testset, y_pred)
mse = mean_squared_error(y_testset, y_pred)
r2 = r2_score(y_testset, y_pred)

# Imprimir resultados
print("MAE:", mae)
print("MSE:", mse)
print("R^2:", r2)
```
