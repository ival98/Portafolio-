```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
link = ("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/drug200.csv")
```

```python
data = pd.read_csv(link)
```

```python
data.head()
```

```python
x = data[["Age", "Sex", "BP", "Cholesterol","Na_to_K"]].values
x[0:5]
```

```python
from sklearn import preprocessing 
prep_sex = preprocessing.LabelEncoder()
prep_sex.fit(["F","M"])
x[:,1] = prep_sex.transform(x[:,1])
```

```python
prep_BP = preprocessing.LabelEncoder()
prep_BP.fit(["LOW", "NORMAL", "HIGH"])
x[:,2] = prep_BP.transform(x[:,2])
```

```python
prep_Ch = preprocessing.LabelEncoder()
prep_Ch.fit(["NORMAL", "HIGH"])
x[:,3] = prep_Ch.transform(x[:,3])
```

```python
y = data["Drug"]
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_trainset, x_testset, y_trainset, y_testset = train_test_split(x,y, test_size=0.2, random_state=3)
```

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth=4)
drugTree
```

```python
drugTree.fit(x_trainset, y_trainset)
```

```python
predictionTree = drugTree.predict(x_testset)
```

```python
print (predictionTree [0:5])
print (y_testset [0:5])
```

```python
from sklearn import metrics
print("Precision de los Arboles de decision: ", metrics.accuracy_score(y_testset, predictionTree))
```
